```python
# define function: measure runtime
def runtime(start, end):
    hour, remainder = divmod(end - start, 3600)
    minute, second = divmod(remainder, 60)
    return 'Runtime: {:>2}:{:0>2}:{:05.2f}'.format(int(hour), int(minute), second)
```
